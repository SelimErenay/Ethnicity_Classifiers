```python
import re, pandas as pd
from pathlib import Path
from collections import Counter

# ====== paths (edit if yours differ) ======
RAW_DIR   = Path("/content/drive/MyDrive/AMAC")
CLEAN_DIR = RAW_DIR / "somali_classifier_splits_v4"   # write cleaned splits here
CLEAN_DIR.mkdir(parents=True, exist_ok=True)

paths = {
    "train": RAW_DIR / "somali_classifier_final_split/train_dataset.csv",
    "eval":  RAW_DIR / "somali_classifier_final_split/eval_dataset.csv",
    "test":  RAW_DIR / "somali_classifier_final_split/test_dataset.csv",
}

# ====== canonicalization helpers ======
LETTERS_ONLY = re.compile(r"[^A-Za-z\s]+")

def tokenize_letters(name: str):
    """letters-only tokens, collapse whitespace, lower."""
    if pd.isna(name): return []
    s = str(name).lower()
    s = LETTERS_ONLY.sub(" ", s)
    s = " ".join(s.split())
    return s.split()

def canonicalize_two_word(name: str, min_len_first=2, min_len_last=2):
    """
    Canonical 'firstname lastname' from any multi-token name.
    Prefer first token length>=min_len_first, last token length>=min_len_last.
    Fall back gracefully if not available (still enforce 2 tokens total).
    """
    toks = tokenize_letters(name)
    if len(toks) < 2:
        return None

    # pick first: earliest token meeting min length, else toks[0]
    first = next((t for t in toks if len(t) >= min_len_first), toks[0])

    # pick last: latest token meeting min length, else toks[-1]
    last = next((t for t in reversed(toks) if len(t) >= min_len_last), toks[-1])

    if not first or not last:
        return None
    return f"{first} {last}"

def normalize_label(x: str) -> str:
    s = str(x).strip().lower()
    if s.startswith("non"): return "non-somali"
    if "somali" in s:       return "somali"
    return "non-somali"

def clean_one(df: pd.DataFrame) -> pd.DataFrame:
    # keep only needed columns if present
    cols = {c.lower(): c for c in df.columns}
    # try to locate name + label columns
    name_col  = cols.get("full_name", cols.get("name", list(df.columns)[0]))
    label_col = cols.get("label", list(df.columns)[1])

    out = df[[name_col, label_col]].copy()
    out.columns = ["full_name", "label"]

    # canonicalize names and labels
    out["full_name"] = out["full_name"].map(canonicalize_two_word)
    out["label"]     = out["label"].map(normalize_label)

    # drop blanks / nans
    out = out.dropna(subset=["full_name", "label"])

    # drop duplicates within the split
    before = len(out)
    out = out.drop_duplicates(subset=["full_name"]).reset_index(drop=True)
    print(f"  - within-split dedup: {before} -> {len(out)}")

    # sanity
    print("  - label counts:", Counter(out["label"]))
    return out

# ====== load & clean each split (individually) ======
raw = {}
for split, p in paths.items():
    print(f"\nðŸ§¹ Cleaning {split}: {p}")
    df = pd.read_csv(p)
    raw[split] = clean_one(df)

# ====== leak check across splits (ensure disjoint canonical names) ======
# remove eval/test rows that collide with train; remove test rows that collide with eval
train_names = set(raw["train"]["full_name"])
eval_names  = set(raw["eval"]["full_name"])

def remove_overlaps(base_names, target_df, tag):
    before = len(target_df)
    target_df = target_df[~target_df["full_name"].isin(base_names)].reset_index(drop=True)
    dropn = before - len(target_df)
    if dropn:
        print(f"  - removed {dropn} overlapping rows from {tag}")
    return target_df

# 1) eval/test vs train
raw["eval"] = remove_overlaps(train_names, raw["eval"], "eval (vs train)")
raw["test"] = remove_overlaps(train_names, raw["test"], "test (vs train)")

# 2) test vs eval
eval_names = set(raw["eval"]["full_name"])
raw["test"] = remove_overlaps(eval_names, raw["test"], "test (vs eval)")

# ====== save cleaned splits ======
out_paths = {
    "train": CLEAN_DIR / "train_dataset.csv",
    "eval":  CLEAN_DIR / "eval_dataset.csv",
    "test":  CLEAN_DIR / "test_dataset.csv",
}
for split, df in raw.items():
    df.to_csv(out_paths[split], index=False)
    print(f"âœ… Saved {split} -> {out_paths[split]} ({len(df)} rows)")
```

```python
# --- helper to sample with a target minority rate (no replacement) ---
def sample_with_target_rate(df, target_rate=0.10, size=None, seed=42):
    rng = np.random.default_rng(seed)
    som = df[df["label"] == "somali"]
    non = df[df["label"] == "non-somali"]

    if size is None:
        size = len(df)

    want_som = int(round(target_rate * size))
    want_non = size - want_som

    if want_som > len(som):
        raise ValueError(f"Not enough Somali rows: want {want_som}, have {len(som)}")
    if want_non > len(non):
        raise ValueError(f"Not enough Non-Somali rows: want {want_non}, have {len(non)}")

    som_idx = rng.choice(som.index.to_numpy(), size=want_som, replace=False)
    non_idx = rng.choice(non.index.to_numpy(), size=want_non, replace=False)
    take = np.concatenate([som_idx, non_idx])

    split_df = df.loc[take].sample(frac=1.0, random_state=seed).reset_index(drop=True)
    remain_df = df.drop(index=take)
    return split_df, remain_df

# ===== NEW SPLIT SETTINGS =====
TEST_FRAC = 0.15        # % of ALL data â†’ test size
EVAL_FRAC = 0.15        # % of ALL data â†’ eval size (same overall share as test)
TARGET_RATE_REAL = 0.10 # 10% Somali for eval & test
RANDOM_STATE = 42
N = len(dedup)

# --- TEST: 10% Somali ---
test_size = int(round(TEST_FRAC * N))
test_df, remain = sample_with_target_rate(
    dedup, target_rate=TARGET_RATE_REAL, size=test_size, seed=RANDOM_STATE
)
print("TEST counts:", test_df["label"].value_counts().to_dict(), "| size:", len(test_df))

# --- EVAL: 10% Somali (from remaining pool) ---
eval_size = int(round(EVAL_FRAC * N))
eval_df, remain = sample_with_target_rate(
    remain, target_rate=TARGET_RATE_REAL, size=eval_size, seed=RANDOM_STATE + 1
)
print("EVAL counts:", eval_df["label"].value_counts().to_dict(), "| size:", len(eval_df))

# --- TRAIN: 50/50 Somali/Non-Somali (downsample the majority) ---
train_pool = remain.copy()
som = train_pool[train_pool["label"] == "somali"]
non = train_pool[train_pool["label"] == "non-somali"]
k = min(len(som), len(non))  # match to the smaller class

train_df = pd.concat([
    som.sample(n=k, random_state=RANDOM_STATE + 2),
    non.sample(n=k, random_state=RANDOM_STATE + 2),
]).sample(frac=1.0, random_state=RANDOM_STATE + 2).reset_index(drop=True)

print("TRAIN counts:", train_df["label"].value_counts().to_dict(), "| size:", len(train_df))

# sanity: show distributions
for name, d in [("train", train_df), ("eval", eval_df), ("test", test_df)]:
    print(f"{name.upper():5} size={len(d):5} | dist={d['label'].value_counts(normalize=True).round(3).to_dict()}")
```

```python
# === ADD: evaluator ===
import re, numpy as np, pandas as pd
from tqdm.auto import tqdm
from sklearn.metrics import classification_report, accuracy_score

def _normalize_label(raw: str) -> str:
    s = raw.strip().lower().replace("-", " ").replace("_", " ")
    if s.startswith("non somali") or ("non" in s and "somali" in s):
        return "Non-Somali"
    if "somali" in s:
        return "Somali"
    return "Non-Somali"

PROMPT_TEMPLATE = """### Instruction:
You will be given a full name. Classify the name into two categories:
1) somali
2) non-somali
Respond with only two words: somali or non-somali

### Input
Name: {name}

### Response
"""

def _make_prompt(name: str) -> str:
    return PROMPT_TEMPLATE.format(name=name)

def evaluate_df(model, tokenizer, df: pd.DataFrame, batch_size=128, max_length=256):
    # Build prompts + gold
    names = df["full_name"].astype(str).tolist()
    gold  = [_normalize_label(x) for x in df["label"].astype(str).tolist()]
    prompts = [_make_prompt(n) for n in names]

    # Generate
    from unsloth import FastLanguageModel
    FastLanguageModel.for_inference(model)

    preds = []
    for i in tqdm(range(0, len(prompts), batch_size), desc="Eval(generation)"):
        batch = prompts[i:i+batch_size]
        toks = tokenizer(
            batch, return_tensors="pt", padding=True, truncation=True,
            max_length=max_length
        ).to(model.device)

        outs = model.generate(
            **toks,
            max_new_tokens=5, do_sample=False, num_beams=1, use_cache=True,
            eos_token_id=tokenizer.eos_token_id, pad_token_id=tokenizer.pad_token_id,
        )

        # decode only the generated suffix
        input_len = toks["input_ids"].shape[1]
        dec = tokenizer.batch_decode(outs[:, input_len:], skip_special_tokens=True)
        preds.extend([_normalize_label(x) for x in dec])

    # Metrics
    acc = accuracy_score(gold, preds)
    report = classification_report(
        gold, preds,
        labels=["Non-Somali", "Somali"],
        target_names=["Non-Somali", "Somali"],
        digits=4,
        zero_division=0,
        output_dict=True,
    )

    # Flatten a few handy keys for early stopping / W&B
    metrics = {
        "accuracy": acc,
        "macro_precision": report["macro avg"]["precision"],
        "macro_recall":    report["macro avg"]["recall"],
        "macro_f1":        report["macro avg"]["f1-score"],
        "weighted_precision": report["weighted avg"]["precision"],
        "weighted_recall":    report["weighted avg"]["recall"],
        "weighted_f1":        report["weighted avg"]["f1-score"],
        # class-wise if you want them visible in W&B too:
        "Somali_precision":     report["Somali"]["precision"],
        "Somali_recall":        report["Somali"]["recall"],
        "Somali_f1":            report["Somali"]["f1-score"],
        "NonSomali_precision":  report["Non-Somali"]["precision"],
        "NonSomali_recall":     report["Non-Somali"]["recall"],
        "NonSomali_f1":         report["Non-Somali"]["f1-score"],
    }
    return metrics, preds, gold, report
```

```python
# === W&B callback ===
import os, random, wandb
from transformers import TrainerCallback
from unsloth import FastLanguageModel

class LLMEarlyStoppingWandbCallback(TrainerCallback):
    def __init__(
        self,
        trainer,                 # <<< pass the trainer object when you build it
        tokenizer,
        eval_df: pd.DataFrame,   # <<< dataframe version of your eval split
        patience: int = 3,
        metric_key: str = "weighted_f1",   # choose: "weighted_f1", "macro_f1", "weighted_recall", etc.
        sample_size: int = 6,
        max_new_tokens: int = 5,
        log_model: str = "checkpoint",
    ):
        self.trainer = trainer
        self.tokenizer = tokenizer
        self.eval_df = eval_df.reset_index(drop=True)
        self.patience = patience
        self.metric_key = metric_key
        self.sample_size = min(sample_size, len(self.eval_df))
        self.max_new_tokens = max_new_tokens
        self.best_metric = -1e9
        self.no_improve = 0
        self.log_model = log_model

        # pick a few samples to show in W&B
        self.sample_rows = (
            self.eval_df.sample(self.sample_size, random_state=42)
            if self.sample_size > 0 else self.eval_df.head(0)
        )

    def on_evaluate(self, args, state, control, **kwargs):
        model = self.trainer.model
        FastLanguageModel.for_inference(model)

        # 1) run our evaluator
        metrics, _, _, report = evaluate_df(
            model=model, tokenizer=self.tokenizer, df=self.eval_df, batch_size=128, max_length=256
        )

        # 2) log to W&B (flatten for nice charts)
        flat = {f"eval/{k}": v for k, v in metrics.items()}
        flat["global_step"] = state.global_step
        wandb.log(flat)

        # pretty print to notebook too (optional)
        print("\nEval metrics:", {k: round(v, 4) for k, v in metrics.items() if isinstance(v, (int, float))})

        # 3) early stopping on chosen metric
        current = float(metrics[self.metric_key])
        if current > self.best_metric:
            self.best_metric = current
            self.no_improve = 0

            # save a "best_<metric>" checkpoint under output_dir/run_name/
            save_dir = os.path.join(args.output_dir, args.run_name, f"best_{self.metric_key}")
            os.makedirs(save_dir, exist_ok=True)
            model.save_pretrained(save_dir)
            self.tokenizer.save_pretrained(save_dir)
            print(f"â¬†ï¸  New best {self.metric_key}: {current:.4f} â€” saved to {save_dir}")
        else:
            self.no_improve += 1
            print(f"â¸ï¸  No improvement on {self.metric_key} ({self.no_improve}/{self.patience})")
            if self.no_improve >= self.patience:
                control.should_training_stop = True
                print(f"ðŸ›‘ Early stopping â€” {self.metric_key} didn't improve for {self.patience} evals.")

        # 4) small sample generations table to W&B
        if self.sample_size > 0:
            table = wandb.Table(columns=["name", "generation"])
            for _, row in self.sample_rows.iterrows():
                prompt = PROMPT_TEMPLATE.format(name=row["full_name"])
                toks = self.tokenizer(
                    prompt, return_tensors="pt", padding=True, truncation=True, max_length=256
                ).to(model.device)

                out = model.generate(
                    **toks, max_new_tokens=self.max_new_tokens, do_sample=False, num_beams=1,
                    eos_token_id=self.tokenizer.eos_token_id, pad_token_id=self.tokenizer.pad_token_id
                )
                gen = self.tokenizer.decode(out[0][toks["input_ids"].shape[1]:], skip_special_tokens=True)
                table.add_data(row["full_name"], gen)
            wandb.log({"eval/sample_predictions": table})

        # go back to train mode
        FastLanguageModel.for_training(model)

        return control
```

```python
#Good function for evaluation for manual runs.

from unsloth import FastLanguageModel
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

def normalize_label(raw: str) -> str:
    """
    Normalize raw labels into exactly 'Somali' or 'Non-Somali'
    """
    s = str(raw).strip().lower().replace("-", " ").replace("_", " ")
    if s.startswith("non somali") or ("non" in s and "somali" in s):
        return "Non-Somali"
    if s.startswith("somali") or "somali" in s:
        return "Somali"
    return "Non-Somali"  # default fallback


PROMPT_TEMPLATE = """### Instruction:
You will be given a full name. Classify the name into two categories:
1) somali
2) non-somali
Respond with only two words: somali or non-somali

### Input
Name: {name}

### Response
"""

def make_prompt(name: str) -> str:
    return PROMPT_TEMPLATE.format(name=name)

def evaluate_on_csv(
    model,
    tokenizer,
    csv_path: str,
    split_name: str = "test",
    batch_size: int = 128,
    max_length: int = 256,
    plot: bool = True,
):
    # 1) Load data
    df = pd.read_csv(csv_path)
    names = df["full_name"].astype(str).tolist()
    gold  = [normalize_label(x) for x in df["label"].astype(str)]
    prompts = [make_prompt(n) for n in names]

    # 2) Generate predictions (batched)
    FastLanguageModel.for_inference(model)  # safe if called multiple times
    preds = []
    for i in tqdm(range(0, len(prompts), batch_size), desc=f"Evaluating ({split_name})"):
        batch = prompts[i:i+batch_size]
        toks = tokenizer(
            batch, return_tensors="pt", padding=True, truncation=True,
            max_length=max_length
        ).to(model.device)

        outs = model.generate(
            **toks,
            max_new_tokens=5,
            do_sample=False,
            num_beams=1,
            use_cache=True,
            eos_token_id=tokenizer.eos_token_id,
            pad_token_id=tokenizer.pad_token_id,
        )

        # only new tokens (exclude the prompt)
        input_len = toks["input_ids"].shape[1]
        gen_only = outs[:, input_len:]

        dec = tokenizer.batch_decode(gen_only, skip_special_tokens=True)
        preds.extend([normalize_label(x) for x in dec])

    # 3) Metrics
    acc = accuracy_score(gold, preds)
    print(f"\nAccuracy ({split_name}): {acc:.4f}\n")

    report = classification_report(
        gold, preds,
        labels=["Non-Somali", "Somali"],
        target_names=["Non-Somali", "Somali"],
        digits=4,
        zero_division=0,
        output_dict=True,   # <-- also return programmatically
    )
    print("Classification report (\033[1m" + split_name + "\033[0m):\n")
    print(pd.DataFrame(report).transpose())

    # 4) Confusion matrix (+ optional plot)
    cm = confusion_matrix(gold, preds, labels=["Non-Somali", "Somali"])
    if plot:
        cm_df = pd.DataFrame(
            cm,
            index=["True Non-Somali", "True Somali"],
            columns=["Pred Non-Somali", "Pred Somali"]
        )
        sns.heatmap(cm_df, annot=True, fmt="d", cbar=False)
        plt.title(f"Confusion Matrix â€” Somali Classifier ({split_name})")
        plt.ylabel("True label"); plt.xlabel("Predicted label")
        plt.tight_layout(); plt.show()

    # 5) Misclassifications table (first 25 preview)
    errors = [(n, g, p) for n, g, p in zip(names, gold, preds) if g != p]
    print("Misclassifications:", len(errors))
    if errors:
        display(pd.DataFrame(errors[:25], columns=["Name", "True", "Pred"]))

    # Return everything you might want later
    return {
        "accuracy": acc,
        "report": report,          # dict form of classification report
        "confusion_matrix": cm,    # numpy array
        "preds": preds,
        "gold": gold,
        "errors": errors,          # list of tuples
    }
```

```python
import os
from transformers import AutoTokenizer
from peft import PeftModel
from unsloth import FastLanguageModel
import seaborn as sns
import matplotlib.pyplot as plt

BASE_4BIT = "unsloth/Meta-Llama-3.1-8B-Instruct-bnb-4bit"
CKPT_PATH = "/content/drive/MyDrive/AMAC/somali_classifier_outputs/somali_classifier_v1.4/best_weighted_f1"

assert os.path.isdir(CKPT_PATH), f"Checkpoint dir not found: {CKPT_PATH}\nls: {os.listdir(os.path.dirname(CKPT_PATH))}"

# Load tokenizer from the checkpoint (keeps PAD/EOS exactly as trained)
tok = AutoTokenizer.from_pretrained(CKPT_PATH, local_files_only=True, use_fast=True)

# Load base model in 4-bit, then attach LoRA adapter from the checkpoint
model, _ = FastLanguageModel.from_pretrained(
    model_name=BASE_4BIT, max_seq_length=256, load_in_4bit=True
)
try:
    model.load_adapter(CKPT_PATH)      # Unsloth helper (if available)
except Exception:
    model = PeftModel.from_pretrained(model, CKPT_PATH, local_files_only=True)

# Now evaluate exactly this saved checkpoint
_ = evaluate_on_csv(
    model=model,
    tokenizer=tok,
    csv_path="/content/drive/MyDrive/AMAC/somali_classifier_final_split_clean/eval_dataset.csv",
    split_name="test",
    batch_size=128,
    max_length=256,
    plot=True,
)
```
